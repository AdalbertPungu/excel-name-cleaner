```python
import pandas as pd

# Charger le fichier Excel (modifiez le nom de fichier au besoin)
df = pd.read_excel('input.xlsx')

def ajuster_noms(row):
    # Convertir en chaînes de caractères et supprimer les espaces inutiles
    first = str(row['First Name']).strip()
    last = str(row['Last Name']).strip()

    # Découper les noms
    first_parts = first.split()
    last_parts = last.split()

    # --- 1️ Si Last Name contient plusieurs mots ---
    if len(last_parts) > 1:
        # On garde uniquement le dernier mot dans Last Name
        dernier_nom = last_parts[-1]
        # Les autres vont à la fin du First Name
        first_parts += last_parts[:-1]
        row['First Name'] = ' '.join(first_parts).strip()
        row['Last Name'] = dernier_nom
        return row  # priorité à la correction de Last Name d'abord

    # --- 2️ Si Last Name est vide et First Name contient 2+ mots ---
    if len(last_parts) == 0 and len(first_parts) >= 2:
        # Déplacer le dernier mot de First Name vers Last Name
        dernier_nom = first_parts.pop(-1)
        row['First Name'] = ' '.join(first_parts).strip()
        row['Last Name'] = dernier_nom
        return row

    # --- 3️ Si Last Name contient déjà un mot unique ---
    if len(last_parts) == 1:
        # Rien à déplacer dans Last Name, mais on peut nettoyer First Name
        row['First Name'] = ' '.join(first_parts).strip()
        row['Last Name'] = last_parts[0].strip()
        return row

    return row

# Appliquer la fonction à chaque ligne
df = df.apply(ajuster_noms, axis=1)

# Sauvegarder le résultat dans un nouveau fichier Excel
df.to_excel('output.xlsx', index=False)

print("Fichier 'output.xlsx' créé avec un seul nom dans la colonne Last Name.")
```
